```python
import sympy as sym

z = sym.symbols('z')
q, EI, L = sym.symbols('q EI L')

w = sym.Function('w')(z)

# Define the differential equation
d_eq = sym.Eq(EI * w.diff(z, 4), -q/9*z)

w = sym.dsolve(d_eq, w).rhs
```

```python
phi = -w.diff(z)
M = EI * phi.diff(z)
V = M.diff(z)
```

```python
eq1 = sym.Eq(w.subs(z, 0), 0)
eq2 = sym.Eq(phi.subs(z, 0), 0)
eq3 = sym.Eq(M.subs(z, L), 0)
eq4 = sym.Eq(V.subs(z, L), 0)

sol = sym.solve([eq1, eq2, eq3, eq4])[0]
```

```python
values = {EI: 60E6, q:12, L:9}
w = w.subs(sol).subs(values)
phi = phi.subs(sol).subs(values)
M = M.subs(sol).subs(values)
V = V.subs(sol).subs(values)
```

```python
sym.plot(w, (z, 0, values[L]), title='Deflection of a Beam', ylabel='w(z)', xlabel='z', show=True)
sym.plot(phi, (z, 0, values[L]), title='Slope of a Beam', ylabel='phi(z)', xlabel='z', show=True)
sym.plot(M, (z, 0, values[L]), title='Bending Moment of a Beam', ylabel='M(z)', xlabel='z', show=True)
sym.plot(V, (z, 0, values[L]), title='Shear Force of a Beam', ylabel='V(z)', xlabel='z', show=True)
```
